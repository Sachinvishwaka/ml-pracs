```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
df=load_iris()
df
```

```python
x=df.data
x
```

```python
y=df.target
y
```

```python
np.isnan(y).sum
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x,y)
```

```python
plt.figure(figsize=(15,10))
from sklearn import tree
tree.plot_tree(dt,filled=True)
```

# DECISION TREE REGRESSION

```python
from sklearn.datasets import load_boston
data=load_boston()
data
```

```python
x1=data.data
x1
```

```python
y1=data.target
y1
```

```python
np.isnan(y1).sum()
```

```python
from sklearn.tree import DecisionTreeRegressor
dt1=tree.DecisionTreeRegressor(random_state=42,max_depth=2)
```

```python
dt1.fit(x1,y1)
```

```python
plt.figure(figsize=(15,10))
from sklearn import tree
tree.plot_tree(dt1,filled=True)
```

```python
viz1=dtreeviz(dt1,x_data=x1,y_data=y1,target_name='PRICE',feature_names=data.feature_names,title='DECISION TREE-Boston Dataset',show_node_labels=True)
viz1
```
